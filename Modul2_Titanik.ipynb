cabin	Cabin number	

embarked	
Port of Embarkation	C = Cherbourg, Q = Queenstown, S = Southampton
Variable 

Notes
pclass: A proxy for socio-economic status (SES)
1st = Upper
2nd = Middle
3rd = Lower

parch: 
The dataset defines family relations in this way...
Parent = mother, father
Child = daughter, son, stepdaughter, stepson
Some children travelled only with a nanny, therefore parch=0 for them.

##### Как быстро и просто посчитать что-либо? https://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.value_counts.html

```python
data_index['Sex'].value_counts()
```

```python
data_index['Embarked'].value_counts()
```

```python
list_of_ports = []

for i in data_index['Embarked'].value_counts():
    list_of_ports.append(i) 
```

```python
print(list_of_ports)
```

```python
list_of_ports[0]
```

```python
for i in list_of_ports:
    print((i/sum(list_of_ports)*100))        
```
